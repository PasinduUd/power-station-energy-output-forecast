```python
import pandas as pd

df_train = pd.read_csv("../input/power-station-hourly-electrical-energy-output/train.csv")
print("Train Dataset :", df_train.shape)
```

```python
X_train = df_train.iloc[:, 1:5]
print("X Train :", X_train.shape)
y_train = df_train.iloc[:, 5:]
print("y Train :", y_train.shape)
```

Refer : [**Multiple Linear Regression**](https://www.javatpoint.com/multiple-linear-regression-in-machine-learning#:~:text=Multiple%20Linear%20Regression%20is%20one,more%20than%20one%20independent%20variable.&text=Example%3A,of%20cylinders%20in%20a%20car)

```python
from sklearn.linear_model import LinearRegression

MLR_model = LinearRegression()  
MLR_model.fit(X_train, y_train) #Fitting the Multiple Linear Regression(MLR) model to the training set
```

```python
df_test = pd.read_csv("../input/power-station-hourly-electrical-energy-output/test.csv")
print("Test Dataset :", df_test.shape)

X_test = df_test.iloc[:, 1:5]
print("X Test :", X_test.shape)
```

```python
y_predictions = MLR_model.predict(X_test) #Predicting the Test set results 
print("y Predictions :", y_predictions.shape)
```

```python
df_predictions = pd.DataFrame(y_predictions, columns = ["EP"])
df_ids = df_test.iloc[:, :1]
df = pd.concat([df_ids, df_predictions], axis=1)
df.to_csv("MLR_predictions.csv")
```

```python
from sklearn.metrics import mean_absolute_error

# Calculate Mean Absolute Error for training data
MAE = mean_absolute_error(y_train, MLR_model.predict(X_train))
print("MAE :", MAE)
```
