```python
import pandas as pd

df_train = pd.read_csv("../input/power-station-hourly-electrical-energy-output/train.csv")
print("Train Dataset :", df_train.shape)
```

```python
X_train = df_train.iloc[:, 1:5]
print("X Train :", X_train.shape)
y_train = df_train.iloc[:, 5:]
print("y Train :", y_train.shape)
```

Refer : https://towardsdatascience.com/7-of-the-most-commonly-used-regression-algorithms-and-how-to-choose-the-right-one-fc3c8890f9e3#:~:text=The%20best%20known%20estimation%20method,Squares%20(RSS)%20becomes%20minimal.

```python
from sklearn.ensemble import RandomForestRegressor

RFR_model = RandomForestRegressor(n_estimators=20, max_depth=10, criterion='mse') 
RFR_model.fit(X_train, y_train)
```

```python
df_test = pd.read_csv("../input/power-station-hourly-electrical-energy-output/test.csv")
print("Test Dataset :", df_test.shape)

X_test = df_test.iloc[:, 1:5]
print("X Test :", X_test.shape)
```

```python
y_predictions = RFR_model.predict(X_test)
print("y Predictions :", y_predictions.shape)
```

```python
df_predictions = pd.DataFrame(y_predictions, columns = ["EP"])
df_ids = df_test.iloc[:, :1]
df = pd.concat([df_ids, df_predictions], axis=1)
df.to_csv("RFR_predictions.csv", index=False)
```

```python
# https://machinelearningmastery.com/regression-metrics-for-machine-learning/#:~:text=There%20are%20three%20error%20metrics,Mean%20Absolute%20Error%20(MAE)
from sklearn.metrics import mean_absolute_error, mean_squared_error

train_predictions = RFR_model.predict(X_train)
MAE = mean_absolute_error(y_train, train_predictions) # Mean Absolute Error
print("MAE :", MAE)
MSE = mean_squared_error(y_train, train_predictions) # Mean Squared Error
print("MSE :", MSE)
RMSE = mean_squared_error(y_train, train_predictions, squared=False) # Root Mean Squared Error
print("RMSE :", RMSE)
```
